### Graph Creation ###
    Divided in 2 parts
    Part-1
        1. It will create DAG of a flow 
        2. It will take only those nodes which are used in the flow by checking Z attribute
        3. Add all JSON porperties to node attributes
   
    Part-2
        1. Remove link in and link out nodes
        2. Maintain original graph node index.
       


```python
def graph_create(x,tab,simulation_ids):
    G=nx.DiGraph()
    link_dict={}

    for i in range(len(x['id'])):
        if x['z'][i]==tab: #take only nodes which are on the flow (ignore components of subflow)
            G.add_node(x['id'][i])
            edge=str(x.wires[i]) if x['type'][i]!='link out' else str(x.links[i])
            src=x['id'][i]
            edge=edge.replace("]",'')
            edge=edge.replace("[",'')
            edge=edge.replace("'",'')
            edge=edge.replace(" ",'')
            node_array=edge.split(",")
            #store link_in and link_out nodes as key and value pairs
            if x['type'][i]=="link out":
                link_dict[src]=node_array[0]
            #remove feedback connections
            if "endloop" in str(x['name'][i]) : 
                node_array.pop(1)

            #add  all edges excluding feedback     
            for node in node_array:
                if node!='' and node not in simulation_ids: #dont connect to sim id
                    G.add_edge(src,node)
            #add node attributes
            G.nodes[x["id"][i]]['name'] = x["name"][i]
            G.nodes[x["id"][i]]['type'] = x["type"][i]
            G.nodes[x["id"][i]]['func'] = x["func"][i]

       
   
    
    # remove link-in and link out nodes and create direct connections     
    for link_out,link_in in link_dict.items():
        for src,_ in list(G.in_edges(link_out)):
            for _,sink in list(G.out_edges(link_in)):
                children=nx.dfs_successors(G,source=src,depth_limit=1)[src]
                for i in range(len(children)): #to deal with index changing
                    G.remove_edge(src,children[i])
                    if children[i]==link_out:
                        children[i]=sink
                    
                for node in children: # adding the children maintaing original index to deal with true and false path
                    G.add_edge(src,node)
                    
    
        G.remove_node(link_in)
        G.remove_node(link_out)

    return G
```

### Graph Linearization
   #### topo_dict
     1. It will make a dictionary of all nodes of the DAG.
     2. Nodes are the keys of the dictionary in a topological order
     3. Values of the dictionary will be children of each node.
  #### travel_seq
    1. First check  if there is discontinuity in flow if yes it would be always for a branch,add a jump to the next node
    2. If not pop the continous node from prev_child so that proper if else can be added in case of branch node                     3. Check if prev_node was a branch node or a loop node if yes,add its jump node-id and cond whether 
    false or truth path is travelled in the graph attribute .
    4. True path or false path is traversed directly can be found with the help of child node index from node-child map.
    5. endloop nodes add its loop node id
    6. finally add present node in comm_list.

```python
def topo_dict(G):
    next_node_top=list(nx.topological_sort(G))
    tree_top={}
    for i in next_node_top:
        for node_pre,node_post in nx.dfs_successors(G,source=i,depth_limit=1).items():
            tree_top[(node_pre)]=node_post
    else:
        tree_top[i]=[] # end node
    return tree_top
```

```python
def travel_seq(G,tree_top):   
    comm_list=[]
    loop=[]
    i=0
    for node in list(tree_top.keys()): 
        if i!=0:
            if node not in prev_child: #checking for discontinuity
                comm_list.append("jump([jump="+str(prev_child[-1])+"])") #unconditional jump
                i=i+1 # increment since one line is added 
            else: #if not pop out the continous child
                ind=list(tree_top[prev_node]).index(node) #index of directly traversed path 0 means true path 1 means false path
                tree_top[prev_node].pop(ind)
                
            name=re.sub(r'_.*$','',G.nodes[prev_node]['name']) #for branch and loop finding where to jump
            if name=="loop" :
                loop.append(prev_node)
                G.nodes[prev_node]["cond"]=ind # maintain whether true or false path is traversed first with help of path index
                G.nodes[prev_node]["jump"]=str(prev_child[0]) #maintain jump location destination
            elif "branch" in name:
                G.nodes[prev_node]["jump"]=str(prev_child[0])
                G.nodes[prev_node]["cond"]=ind
                
        #for endloop jump location needed
        if "endloop" in G.nodes[node]['name']: #if prev node was a branch
                G.nodes[node]["jump"]=str(loop.pop())
                
        #travel the present node by default
        comm_list.append(str(node))
        prev_node=node
        prev_child=tree_top[node]
        i=i+1
    
    return comm_list,i
```

## IR conversion-1
    1. Convert the node-type to Intermediate Representation
    2. syntax of the type {node_type}([args])
    3. arguments are in Key-Value Pair.
    4. extract arguments from rule-book
    5. comments are written to increase readability.
    

```python
def IR_1(graph,df1,node_dump,prev_line_count):
    line_count=prev_line_count #starts with previous flow line number
    addr_dict={} #stores node-line mapping
    IR_list=[] 
    variable_dict={} #dict object of arguments
    for i in range(len(node_dump)):
        node=(node_dump[i].split(" ")).pop(0)
        if "jump"  in node :
            IR_list.append(node)
            line_count+=1
        else:
            op=[]
            name=graph.nodes[node]['name']
            args=df1[name]["arguments"] #fetch the required argument attributes
            for arg in args:
                if "node" in arg: #if node generated extract attribute values from function atrribute
                    func=str(graph.nodes[node]['func']).split("\n")
                    attr=re.sub(r'^.*node.','',arg)
                    match=re.search(attr,func[0])
                    if match:
                        var=re.sub(r'^.*'+str(attr)+'=','',func[0])
                        var=var.split(";")[0]
                        if "global" in var:
                            var=re.sub(r'^.*global.','',var)
                        elif "msg" in var:
                            var=re.sub(r'^.*msg.','',var)
                            var="msg_"+var
                        variable_dict[attr]=var
                        var=str(attr)+"="+str(var)
                        op.append(var)            
                elif "graph" in arg: #if graph traversal generated directly find attribute from graph
                    attr=re.sub(r'^.*graph.','',arg)
                    var=graph.nodes[node][attr]
                    variable_dict[attr]=var
                    var=str(attr)+"="+str(var)
                    op.append(var)       
            comment=df1[name]['comment'].format(**variable_dict) #fetch the comment from rule book change its' dynamic content
            IR_list.append(str("//"+comment+"\n"+name+"("+str(op)+")"))
            addr_dict[node]=line_count #map nodes to line numbers
            line_count+=1
     
        
    return IR_list,addr_dict,line_count
```

### Creation of Variable Table:
    
    1. For all initialization nodes ,it will make an entry to the variable_hash along with it's size.
    3. If the node is of macro(global) node ,store it in the macro hash which will help us to define include #def statements (macros)
      

```python
def hash_create(IR_local):
    hash_tab={}
    macro_hash={}
   
    for i in range(len(IR_local)):
        ir=IR_local[i].split("\n").pop()
        comm=re.sub(r'\(.*$','',ir)
        comm=re.sub(r'^.*: ',"",comm)
        arguments=re.sub(r'^.*\(','',ir)
        
        arguments=arguments.replace(")","")
        arguments=arguments.replace("'","")
        arguments=arguments.replace("[","")
        arguments=arguments.replace("]","")
        arguments=arguments.replace('"',"")
        arguments=arguments.replace(' ',"")

        
        if "init" in comm:
            var_name=re.sub(r'^.*init=','',arguments).split(",").pop(0) 
            var_size=re.sub(r'^.*size=','',arguments).split(",").pop(0)     
            hash_tab[var_name]=var_size
        elif "macros" in comm:
            var_name=re.sub(r'^.*init=','',arguments).split(",").pop(0) 
            var_size=re.sub(r'^.*size=','',arguments).split(",").pop(0) 
            var_val=re.sub(r'^.*val=','',arguments).split(",").pop(0)  
            macro_hash[var_name]={}
            macro_hash[var_name]['size']=var_size
            macro_hash[var_name]['val']=var_val
        
    return hash_tab,macro_hash
```

### LiveRange Analysis:
    1. We need to find live ranges of each variables to allocate registers with the help of the graph.
    Part-A
        1. Find the variables which are used and written by a node.
        2. Maintain them in two different dictionaries
    Part-B:
        1. Find live points of the variables.
        2. Take graph object (control flow graph) into consideration.
        3. Feedback connection needs to be added so that exact control flow can be known 
        3. A Variable is live at a point if it is used by the node OR {it is live at it's children AND not written by the node}
        4. Live[node]=(Live[Children]-Def[node])U(use[node])
    Part-C
        1. convert Live point to live ranges.
        2. Start with each definition point and move in the graph in DFS.
        3. end the definition when variable is live at a node but not live at it's all children
        4. Merge the overlapping ranges


```python
##use and definition dictionary to extarct uses and definitions of each node
u_d={
    'loop_for':{
        'use':["op1",'op2'],
        'def':[]
    },
    'loop_while':{
        'use':["op1",'op2'],
        'def':[]
    },
    'branch':{
        'use':["op1",'op2'],
        'def':[]
    },
    'eval':{
        'use':["op1","op2"],
        'def':['store']
    },
    "delay":{
        'use':[],
        'def':[]
    },
    "init":{
        'use':[],
        'def':[]
    },
    "macros":{
        'use':[],
        'def':[]
    },
    "input":{
        "use":[],
        'def':[]
    },
    'output':{
        'use':["read"],
        "def":[]
    },
    'subflow':{
        'use':[],
        "def":[]
    },
    "copy":{
        'use':['read'],
        'def':['write']
    },
    'endloop_for':{'use':["iter"],
                'def':["iter"]
                  },
    "endloop_while":{
        'use':[],
        'def':[]
    },
    'end_if':{
        'use':[],
        'def':[]
    },
    'end':{
        'use':[],
        'def':[]
    }
}
```

```python
 ## function to convert live points to live ranges
##var-->variable for which live range is found,
##node--> node where you are checking live or not
## graph--> graph object of that flow
##out_local-->live variables dictionary
##node-line-->node-line mapping
def convert_liveRange(var,node,graph,out_local,node_line):
    end=[]
    out_list=list(out_local.keys())
    if var not in out_local[node_line[node]]: #if variable not in live at the node return
        return [node_line[node]]
                
    else: # if it is live at the node run a dfs for the children 
        try:
            children=nx.dfs_successors(graph,source=node,depth_limit=1)[node]
        except KeyError:
            return [node_line[node]]
        for child in children:
            end_line=convert_liveRange(var,child,graph,out_local,node_line) ##recursion to find end point
            end_line=[j for j in end_line if j not in end]
            end+=end_line
    return end #return end points. 
```

```python
##liveness analysis function
#inputs graph dictionary, flow-starting line, IR code, Node-line mapping, use-def map of each IR, variable_hash
def liveness(graph_dict,flow,IR,node_line,u_d_map,variable_hash):  
    graph=graph_dict[flow] ## take the graph object of the respective flow
    node_child=topo_dict(graph) ## get the node-child map of the graph
    dict_key=list(node_child.keys())
    in_={} #dict mapping IR-lines and live variables
    out_={} #
    use_={} ##dict mapping IR-lines and variables used
    def_={} ##dict mapping IR-lines and variables written
    push_list={} #maintains a list of parameters passed to the subflow
    pop_list={} #maintains a list of parameters returned by a subflow
    live_range={}
    for node in node_child.keys():
        in_[node_line[node]]=set()
        out_[node_line[node]]=set()
        use_[node_line[node]]=set()
        def_[node_line[node]]=set()
    ##part-1 find u-d
    for node in node_child.keys():
        ir=IR[node_line[node]]
        ir=ir.split("\n").pop()
        ir=re.sub(r'^.*: ','',ir)
        comm=re.sub(r'\(.*$','',ir)
        args=re.sub(r'^.*\(','',ir)
        args=args.replace(")","")
        args=args.replace("'","")
        args=args.replace("[","")
        args=args.replace("]","")
        args=args.replace('"',"")
        args=args.replace(' ',"")
        def1=[]
        use1=[]
        for attr in u_d_map[comm]['use']:
            var=re.sub(r'^.*'+attr+"=",'',args)
            var=var.split(',')[0]
            use1.append(var)
            for attr in u_d_map[comm]['def']:
                var=re.sub(r'^.*'+attr+"=",'',args)
                var=var.split(',')[0]
                def1.append(var)
            def1=[j for j in def1 if j in variable_hash.keys()]
            use1=[j for j in use1 if j in variable_hash.keys()]
            def_[node_line[node]]=def_[node_line[node]].union(set(def1))
            use_[node_line[node]]=use_[node_line[node]].union(set(use1)) 
                

         
    ##find live point (part-2)
    change=True   
    j=0
    loop={} 
    loop_start=[]
    for node in dict_key: #add feedback connection again. Needed to build exact CFG
        ir=IR[node_line[node]]
        ir=ir.split("\n").pop()
        ir=re.sub(r'^.*: ','',ir)
        name=re.sub(r'\(.*$','',ir)
        if name=='loop_for' or name=="loop_while":
            loop_start.append(node)
        if 'endloop' in name:
            start=loop_start.pop()
            loop[node]=start
    for end in loop.keys():
        node_child[end].append(loop[end])
    # Finding the live points of the variable
    while change: #repeat till it converges
        count=0
        for node in node_child.keys():
            children=node_child[node]
            ot=out_[node_line[node]]
            i=in_[node_line[node]]
            for child in children:
                out_[node_line[node]]=out_[node_line[node]].union(in_[node_line[child]]) #live variable at output point(union of live of alll children)
            in_[node_line[node]]=use_[node_line[node]].union(out_[node_line[node]]-def_[node_line[node]])#live at input point of a node
            if j!=0: ##repeat until converges
                if i!=in_[node_line[node]]:
                    count+=1
                if ot!=out_[node_line[node]]: count+=1
            else:
                count=1
        if count==0:
            break

        j+=1
        
    #find live ranges
    in_list=list(in_.keys())
    for var in variable_hash.keys():
        
        var_type=var
        def_points=[]
        
        if var_type in in_[int(flow)]: #take care of variables which are live without definition (in case of subflows it is possible)
            def_node=[j for j in node_line.keys() if node_line[j]==int(flow)].pop()
            def_points.append(def_node)
        for node in dict_key: #LIST Of definition points
            if var_type in def_[node_line[node]]:
                def_points.append(node)

        for i in range(len(def_points)): #start with each defnition points
                def_node=node_line.keys()
                end=convert_liveRange(var_type,def_points[i],graph,out_,node_line)
                for end_line in end: ## store in a tuple (start,end, flow_start_line)
                    try:
                        live_range[var_type].append((node_line[def_points[i]],end_line,flow))
                    except KeyError :
                        live_range[var_type]=[(node_line[def_points[i]],end_line,flow)]

    return use_,def_,in_,out_,live_range,push_list,pop_list
```

### Register Allocation 
    1. Allocate register based on live ranges.(holes will be considered)
    2. Bank will be allocated depending upon maximum usage in that range
    3. If register is not available memory will be allocated.
    4. 2-registers will be kept un-allocated (R0,R1) for indexed addressing for main flow.
    5.memory locations-->22h-50h(47 bytes); bit locations-->20h-21h(16bits); stack location-->51h-7Fh(47 bytes)

```python
##  sort the range in increasing order of starting point
ranges_var=[]
live=new_live
for var in live.keys():
    for start,end,flow in live[var]:
        ranges_var.append((start,end,var,flow))

ranges_var.sort()        
ranges_var
```

```python
## register allocation
regs_bank=[list(range(2,8)),list(range(8)),list(range(8)),list(range(8))] ##register list for main flow
mem_addr=list(range(34,81)) #available memory address (22h to 50h)
bit_addr=list(range(16)) ##bit addressable area 00h to 0Fh

active_ranges=[] ##list of active ranges
mem_alloc={} # variables in memory
bit_alloc={} #variables in bit area
reg_des={} #variables in registers
reg_bank={} #flow-wise number of free registers.
for intr in ranges_var:
    bank_count=[0]
    pop_active=[]
    var_type=int(get_type(intr[2],variable_hash,macro_hash)) #get the type of the variable
    flow_line=intr[3] ##get the flow line
    if flow_line not in reg_bank.keys(): 
        if flow_line=='0':#in case of main flow
            reg_bank[flow_line]=[list(range(2,8)),list(range(8)),list(range(8)),list(range(8))]
        else: #in case of subflows keep R2,R3 free for storing return address
            reg_bank[flow_line]=[list(range(4,8)),list(range(8)),list(range(8)),list(range(8))]

    for active_intr in active_ranges:#if active range expired add the reg to free list
        reg,bank=reg_des[active_intr].split("_")
        bank=int(bank)
        bank_count[bank]+=1
        if active_intr[1]<=intr[0]:
            reg_bank[active_intr[3]][bank].append(int(reg))
            reg_bank[active_intr[3]][bank].sort() #keep the list sorted so that maximum consecutive address available
            pop_active.append(active_intr) 
  
    active_ranges=[j for j in active_ranges if j not in pop_active]        #remove expired interval from active list
    if (var_type/8)>1:
            pass #replace with multi byte variable algorithm
    ## deal with 8 bit vars
    elif (var_type)==8: ##8 but vars
        for i in range(len(regs_bank)):
            tent_allocated_bank=bank_count.index(max(bank_count)) #find bank that used max
            if len(reg_bank[flow_line][tent_allocated_bank])>0:
                alloc_reg=reg_bank[flow_line][tent_allocated_bank].pop(0)
                reg_des[intr]=str(alloc_reg)+"_"+str(tent_allocated_bank) #store regName_bank
                active_ranges.append(intr)
                break
            else:
                bank_count[tent_allocated_bank]=-1
        else: #spill(if memory locations are not enough)
            mem_alloc[intr]=str(mem_addr.pop(0))
            
    elif (var_type)==1: #bit variables
        bit_alloc[intr]=str(bit_addr.pop(0))
```

### Final Assembly Conversion (8051)
    1. Making pass through the IR find it's matching instruction.
    2. Replace the variables with allocated registers in that range.
    3. Remove end-if and init nodes.
    4. Map macro constants to asm macros
   #### Simplifications:
       1. Different addressing modes are not considered
       2. Indexed (array) addressing mode is not implemented
       3. Only register addressing mode and direct(in case of memory location) is considered
       4. Some copy and jump lines are redundant which is not optimized
     
    

```python
#function to get the register allocated to the variable --> used when converting to final assembly
def get_reg(reg_des_local,mem_local,bit_local,line_no,instruction):
        lines=instruction.split("\n")
        new_ins=""
        for line in lines:
            operands=line.split(" ").pop()
            operands=operands.split(",")
            for var in operands:
                    line=line.replace(var,get_hex(var)) # get hex value of numbers
                    if var in reg_des_local.keys():
                        for start,end in reg_des_local[var].keys(): #if var is in  reg find which reg is allocated in that range
                            if line_no>=start and line_no<=end:
                                reg=reg_des_local[var][(start,end)]
                                line=line.replace(var,"R"+reg)
                                break
                        
                    elif var in mem_local.keys():
                        for start,end in mem_local[var].keys(): #if var is in  mem find which reg is allocated in that range
#                           
                            if line_no>=start and line_no<=end:
                                reg=mem_local[var][(start,end)]
                                reg=get_hex(reg)[1:]
                                line=line.replace(var,reg)
                                break
                    elif var in bit_local.keys():
                        for start,end in bit_local[var].keys():#if bit is in  mem find which reg is allocated in that range
                            if line_no>=start and line_no<=end:
                                reg=bit_local[var][(start,end)]
                                reg=get_hex(reg)[1:]
                                line=line.replace(var,reg)
                                break
                    
                        
            new_ins+=line+"\n"
        return new_ins.strip()
```

```python
## returns hex number of a decimal number
def get_hex(string):
    try:
        num_int=int(string)
        hex_int="{:02X}".format(num_int & ((1 << 8)-1))
        hex_int="#"+hex_int+"h"
        return hex_int
    except ValueError:
        return string
```

```python
##functions to map to proper assembly
def ALU(arg_dict): # all alu operations
    case=arg_dict['operand']
    instruction={"+":"mov A,{op1}\nadd A,{op2}\nmov {store},A",
                "-":"mov A,{op1}\nclr C\nsubb A,{op2}\nmov {store},A",
                "~":"mov {store},/{op2}",
                }
    return instruction[case].format(**arg_dict)


def cond_jmp(arg_dict,var_type): #condtional jumps
    if var_type=='8': #byte variables
        case=arg_dict['relop']+"_"+str(arg_dict['cond'])
        # negiate condition depending upon whether true or false path is traversed first
        if case=="eq_0": inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJNZ {jump}"
        elif case=="eq_1": inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJZ {jump}"

        elif case=="neq_0": inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJZ {jump}"
        elif case=="neq_1": inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJNZ {jump}"
        
        elif case=="lte_0" : inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJNC {jump}" #jump if gt
        elif case=="lte_1": inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJC {jump}\nJZ {jump}" #jump is lte
        elif case=="gte_0" : inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJC {jump}" # jump if lt
        elif case=="gte_1": inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJNC {jump}\nJZ {jump}" #jump if gte
        elif case=="lt_0" : inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJNC {jump}\nJZ {jump}" #jump if gte
        elif case=="lt_1" : inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJC {jump}" #jump if lt
        elif case=="gt_0" : inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJC {jump}\nJZ {jump}" #jump if lte
        elif case=="gt_1" : inst="mov A,{op1}\nclr C\nsubb A,{op2}\nJNC {jump}" #jump if lt
    elif var_type=='1': #bit variables
        
        case=arg_dict['relop']+"_"+str(arg_dict['cond'])+"_"+arg_dict['op2']
        if case=="eq_0_true": inst="JNB {op1},{jump}"
        elif case=="eq_0_false": inst="JB {op1},{jump}"
        elif case=="eq_1_true": inst="JB {op1},{jump}"
        elif case=="eq_1_false": inst="JNB {op1},{jump}"
        elif case=="neq_0_true": inst="JB {op1},{jump}"
        elif case=="neq_0_false": inst="JNB {op1},{jump}"
        elif case=="neq_1_true": inst="JNB {op1},{jump}"
        elif case=="neq_1_false": inst="JB {op1},{jump}"
    
    return inst.format(**arg_dict)

def mov(arg_dict,var_type): #copy instruction
    if var_type=='8': #8 bit copy
        inst_str="mov A,{read}\nmov {write},A"
    elif var_type=='1': # single bit true and false condtions
        if arg_dict['read']=='false':
            inst_str="clr {write}"
        elif arg_dict['read']=='true':
             inst_str="setb {write}"
        else: # bit moves
            inst_str="mov C,{read}\nmov{write},C"
    
    return inst_str.format(**arg_dict)

def jump(arg_dict): #uncondtional jump
    inst="JMP {jump}".format(**arg_dict)
    return inst


def call(arg_dict,push_list,pop_list): #call instruction
    subflow_line=arg_dict['call']
    subflow_line=re.sub(r'^.*-','',subflow_line)
    inst=''
    for params in push_list[subflow_line]: #push the required parameters (can be known from live ness analysis)
        inst+="mov A,"+params+"\nPush E0h\n"
    inst+="call {call}\n".format(**arg_dict)
    for params in pop_list[subflow_line]: #pop the returned parameter (can be knwon from live-ness analysis)
        inst+="pop E0h\n"+"mov "+params+",A\n"
    return inst
```

```python

flow=[]
flow_mapp={}
line_count=0
subflow_list=[]
for i in range(len(IR_2)):
        arg_dict={}
        ir=IR_2[i].split("\n").pop()
        comm=re.sub(r'\(.*$','',ir)
        comm=re.sub(r'^.*: ',"",comm)
        args=re.sub(r'^.*\(','',ir)
        args=args.replace(")","")
        args=args.replace("'","")
        args=args.replace("[","")
        args=args.replace("]","")
        args=args.replace('"',"")
        args=args.replace(' ',"")
        args=args.split(",")
        ##direct mappings of IRs to instructions
        for arg in args:
            value=re.sub(r'^.*=','',arg)
            key=re.sub(r'=.*$','',arg)
            arg_dict[key]=value
        if comm=="loop_while":
            op_type=get_type(arg_dict['op1'],variable_hash,macro_hash)
            ins=cond_jmp(arg_dict,op_type)
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            flow_mapp[i]=line_count
            ins=ins.split('\n')
            for line in ins:
                line_count+=1
                flow.append(line)

        elif comm=="loop_for":
            op_type=get_type(arg_dict['op1'],variable_hash,macro_hash)
            ins=cond_jmp(arg_dict,op_type)
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            flow_mapp[i]=line_count
            ins=ins.split('\n')
            for line in ins:
                line_count+=1
                flow.append(line)
        elif comm=="eval":
            
            ins=ALU(arg_dict)
            print(ins)
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            ins=ins.strip()
            flow_mapp[i]=line_count
            ins=ins.split('\n')
            for line in ins:

                line_count+=1
                flow.append(line)
        elif comm=="end":
            ins=';end'
            flow_mapp[i]=line_count
            flow.append(ins)
            line_count+=1
        elif comm=="copy":
            op_type=get_type(arg_dict['write'],variable_hash,macro_hash)
#             print(op_type)
            ins=mov(arg_dict,op_type)
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            flow_mapp[i]=line_count
            ins=ins.split('\n')
            for line in ins:
                line_count+=1
                flow.append(line)
            
        elif comm=="branch":
            op_type=get_type(arg_dict['op1'],variable_hash,macro_hash)
            ins=cond_jmp(arg_dict,op_type)
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            flow_mapp[i]=line_count
            ins=ins.split('\n')
            for line in ins:
                line_count+=1
                flow.append(line)
            
        elif comm=="subflow":
            ins=call(arg_dict,push_param,pop_param)
            subflow_no=[re.sub(r'^.*-','',arg_dict['call'])]
            subflow_list.extend([j for j in subflow_no if j not in subflow_list ])
            pus_ins=""
            pop_ins=""
            for var in reg_des2.keys():
                for start,end in reg_des2[var]:
                    if i>=start and i<=end:
                        pus_ins+="mov A,"+var+"\npush E0h\n"
                        pop_ins+="pop E0h\nmov "+var+",A\n"
            ins=pus_ins+ins+pop_ins
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            ins=ins.strip()
            flow_mapp[i]=line_count
            ins=ins.split('\n')
            for line in ins:
                line_count+=1
                flow.append(line)
            
        elif comm=="jump":
            ins=jump(arg_dict)
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            flow_mapp[i]=line_count
            ins=ins.split('\n')
            for line in ins:
                line_count+=1
                flow.append(line)
            
        elif comm=="endloop_for":
            ins='INC '+arg_dict['iter']
            ins+="\n"+jump(arg_dict)
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            ins=ins.split('\n')
            for line in ins:
                line_count+=1
                flow.append(line)
            flow_mapp[i]=line_count
        elif comm=="output":
            op_type=get_type(arg_dict['write'],variable_hash,macro_hash)
            ins=mov(arg_dict,op_type)
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            flow_mapp[i]=line_count
            ins=ins.split('\n')
            print(ins)
            for line in ins:
                line_count+=1
                flow.append(line)
            
        elif comm=="endloop_while":
            ins=jump(arg_dict)
            ins=get_reg(reg_des2,mem_alloc2,bit_alloc2,i,ins)
            ins=ins.split('\n')
            for line in ins:
                line_count+=1
                flow.append(line)
            flow_mapp[i]=line_count
            
        elif comm=="RET":
            flow.append(comm)
            flow_mapp[i]=line_count
            line_count+=1
        elif comm=="delay":
            flow.append("delay")
            flow_mapp[i]=line_count
            line_count+=1
        else:
            flow_mapp[i]=line_count
flow
    
```
